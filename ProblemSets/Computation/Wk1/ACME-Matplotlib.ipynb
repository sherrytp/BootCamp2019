### Problem 1

````python
def sample(n): 
    array = np.random.normal(size = (n,n))
    arr_mean = array.mean(axis=1)
    return (np.var(arr_mean))
````

````python
def plot_sample(): 
    plt.ion()   # Turn interactive mode on and make some plots. 
    n = np.arange(100, 1101, 100)
    arr = []
    for i in range(len(n)): 
        arr.append(sample(n[i]))
    plt.plot(arr)
    plt.show()
````

````python
plot_sample()

plt.clf()   # Clear the figure 
plt.ioff()   # turn interactive mode off. 
````

### Problem 3

````python
def f(x): 
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/(x-1)

x1 = np.linspace(-2,1, 50)
y1 = f(x1)
x2 = np.linspace(1,6, 50)
y2 = f(x2)

plt.plot(x1, y1, 'c--', linewidth=4)
plt.plot(x2, y2, 'c--', linewidth=4)
plt.xlim(-2,6)
plt.ylim(-6,6)
plt.show()
````

### Problem 4

````python
ax1 = plt.subplot(221)
ax2 = plt.subplot(222)
ax3 = plt.subplot(223)
ax4 = plt.subplot(224)

ax1.axis([0, 2*np.pi, -2, 2])
ax2.axis([0, 2*np.pi, -2, 2])
ax3.axis([0, 2*np.pi, -2, 2])
ax4.axis([0, 2*np.pi, -2, 2])

x = np.linspace(0, 2*np.pi, 50)
ax1.set_title('$sin(x)$')
ax2.set_title('$sin(2x)$')
ax3.set_title('$2sin(x)$')
ax4.set_title('$2sin(2x)$')

ax1.plot(x, np.sin(x), 'g-')
ax2.plot(x, np.sin(2*x), 'r--')
ax3.plot(x, 2*np.sin(x), 'b--')
ax4.plot(x, 2*np.sin(2*x), 'm:')

plt.suptitle("Graphs")     # also fig.suptitle()
````

1. plt.axes() creates a new place to draw on the figure, while 
```plt.axis() (or ax.axis())
```
sets properties of the x- and y-axis in the current axes, such as the x and y limits.
2. ```plt.subplot()``` (singular) returns a single subplot belonging to the current figure, while
```plt.subplots()``` (plural) creates a new figure and adds a collection of subplots to it.

### Heatmap & Contour map

````python
x = np.linspace(-np.pi, np.pi, 100)
y = x.copy()
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)    # g(x,y) = sin(x)sin(y)

plt.subplot(131)
plt.pcolormesh(X,Y,Z, cmap = "viridis")
plt.colorbar()
plt.xlim(-np.pi, np.pi)
plt.ylim(-np.pi, np.pi)

plt.subplot(132)
plt.contour(X,Y,Z, 10, cmap = "coolwarm")
plt.colorbar()

plt.subplot(133)
plt.contourf(X,Y,Z, [-1, -0.8, -.5, 0, .5, .8, 1], cmap = "magma")
plt.colorbar()
plt.show
````

### Problem 6

````python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = x.copy()
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y) /(X*Y)

plt.subplot(121)
plt.pcolormesh(X,Y,Z, cmap = "viridis")
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.colorbar()

plt.subplot(122)
plt.contour(X,Y,Z, 10, cmap = "coolwarm")
plt.axis([-2*np.pi, 2*np.pi, -2*np.pi, 2*np.pi])
plt.colorbar()

plt.show
````

### 3D Plotting 

````python
x = np.linspace(-np.pi, np.pi, 200)
y = np.copy(x)
X, Y = np.meshgrid(x, y)
Z = np.sin(X) * np.sin(Y)    # g(x) = sin(x)sin(y)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(1,1,1, projection = '3d')
ax.plot_surface(X,Y,Z)
plt.show()
````

### Animation

````python
from matplotlib.animation import FuncAnimation 

def sine_animation(): 
    x = np.linsapce(0, 2*np.pi, 200)[:-1]
    y = np.sin(x)
    
    fig = plt.figure()
    plt.xlim(0, 2*np.pi)
    plt.ylim(-1.2, 1.2)
    
    drawing, = plt.plot([], [])   # draw an empty line, COMMA!! 
    
    def update(index): 
        drawing.set_data(x[:index], y[:index])
        return drawing,     # COMMA! 
    
    a = FuncAnimation(fig, update, frames=len(x), interval=10)
    plt.show()
````
